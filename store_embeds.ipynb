```python
import pandas as pd
```

```python
data = pd.read_csv("Fashion_Dataset.csv")
data
```

```python
pro_name = list(data["name"])
pro_desc = list(data["description"])
pro_img = list(data["img"])
pro_color = list(data["colour"])
pro_price = list(data["price"])
```

```python
chunks = []
metadata = []

for i in range(len(pro_desc)):
    temp = str(pro_name[i]) + "." + str(pro_desc[i]) + "." + str(pro_color[i])
    chunks.append(temp)
    metadata.append({
        "color" : str(pro_color[i]).strip().lower(),
        "price" : pro_price[i],
        "img_url" : pro_img[i],
        "name" : pro_name[i]
    })

chunks[0]
metadata[0]
```

```python
!pip install sentence_transformers
!pip install chromadb
```

```python
import chromadb
import os
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-mpnet-base-v2")
embeds = model.encode(chunks)
```

```python
# client.delete_collection("clothes_data")
```

```python
client = chromadb.PersistentClient(path = os.getcwd())
collection = client.get_or_create_collection("clothes_data")
```

```python
collection.add(
    metadatas = metadata,
    embeddings = embeds,
    ids = [str(i) for i in range(len(chunks))]
)
```
